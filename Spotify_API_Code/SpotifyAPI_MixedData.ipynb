## Genre Encoded Mapping:
 $$ 0 \to \text{Rap}\ , 1 \to \text{Country} \ ,2 \to \text{Pop} \ ,3 \to \text{Jazz} $$

```python
rap = pd.read_csv("/content/drive/MyDrive/Spotify/Spotify_API_csv/SpotifyAPI_Rap.csv")
genre = [0 for i in range(rap.shape[0])]
rap.insert(loc=3, column='genre', value=genre)

country = pd.read_csv("/content/drive/MyDrive/Spotify/Spotify_API_csv/SpotifyAPI_Country.csv")
genre = [1 for i in range(country.shape[0])]
country.insert(loc=3, column='genre', value=genre)

pop = pd.read_csv("/content/drive/MyDrive/Spotify/Spotify_API_csv/SpotifyAPI_Pop.csv")
genre = [2 for i in range(pop.shape[0])]
pop.insert(loc=3, column='genre', value=genre)

jazz = pd.read_csv("/content/drive/MyDrive/Spotify/Spotify_API_csv/SpotifyAPI_Jazz.csv")
genre = [3 for i in range(jazz.shape[0])]
jazz.insert(loc=3, column='genre', value=genre)

mix = pd.concat([rap, country, pop, jazz], ignore_index=True).sample(frac=1).reset_index() # Concatonates and shuffles the order
mix.reset_index() # resets index
mix = mix.drop(['Unnamed: 0', 'index'], axis=1)

mix.to_csv('SpotifyAPI_AllGenres.csv')
!cp SpotifyAPI_AllGenres.csv "drive/My Drive/Spotify/"
```
