```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import billboard
import datetime
cid = 'cid'
secret = 'secret'
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager
=
client_credentials_manager)
```

```python
song_data = []


for i in range(0,1000,50):
  query = "year:2021-2022 genre:jazz"
  results = sp.search(q=query, type="track", limit=50, offset=i)
  tracks = results["tracks"]["items"]
  for track in tracks:
    name = track["name"][:track["name"].index(' [with')] if ' [with' in track["name"] else track["name"]
    name = name[:name.index(' [feat')] if ' [feat' in name else name
    name = name[:name.index(' (with')] if ' (with' in name else name
    name = name[:name.index(' (feat')] if ' (feat' in name else name
    artist = track["artists"][0]["name"]
    features = sp.audio_features(track["id"])[0]
    song_data += [{
        "name": name.upper(),
        "artist": artist.upper(),
        "danceability": features["danceability"],
        "energy": features["energy"],
        "key": features["key"],
        "loudness": features["loudness"],
        "mode": features["mode"],
        "speechiness": features["speechiness"],
        "acousticness": features["acousticness"],
        "instrumentalness": features["instrumentalness"],
        "liveness": features["liveness"],
        "valence": features["valence"],
        "tempo": features["tempo"],
        "duration_ms": features["duration_ms"]
    }]

songs_df = pd.DataFrame(song_data)

start_date = datetime.date(2021, 1, 1)
end_date = datetime.date(2022, 12, 31)

current_date = start_date
hit_songs = set()

while current_date <= end_date:
    chart = billboard.ChartData("jazz-songs", date=current_date)
    for song in chart:
      title = song.title
      artist = song.artist[:song.artist.index(" Featuring")]  if " Featuring" in song.artist else song.artist
      artist = artist[:artist.index(" &")]  if " &" in artist else artist
      artist = artist[:artist.index(",")]  if "," in artist else artist
      hit_songs.update([(title.upper(), artist.upper())])

    # print(current_date, hit_songs)
    # hit_songs.update([(song.title, song.artist[:song.artist.index(" Featuring")]  if " Featuring" in song.artist else song.artist) for song in chart])
    current_date += datetime.timedelta(weeks=1)
    # print(current_date, [(song.title, song.artist[:song.artist.index(" Featuring")]  if " Featuring" in song.artist else song.artist) for song in chart])

# Assign "hit" feature in the DataFrame
songs_df["hit"] = songs_df.apply(lambda row: 1 if (row["name"], row["artist"]) in hit_songs else 0, axis=1)

songs_df

from google.colab import drive
drive.mount('/content/drive')

songs_df.to_csv('SpotifyAPI_jazz21-22.csv')
!cp SpotifyAPI_jazz21-22.csv "drive/My Drive/Spotify/"
```
